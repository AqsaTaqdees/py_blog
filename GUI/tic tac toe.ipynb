```python
from tkinter import *
import random
```

```python
def next_turn(row,column):
    global player
    #if the butotn we click on is empty and to access the text of button add ['text'] here. Here False means there is no winner
    if buttons[row][column]['text'] == "" and check_winner() is False:
        if player == players[0]:
            buttons[row][column]['text'] = player
            
            #if there is no winner in the previous game; swap the player with the next player
            if check_winner() is False:
                player = players[1]
                #here we are configuring our labels so that it displays the next player's turn i.e. player at index 1's turn
                label.config(text=(players[1]+" turn"))
                
            elif check_winner() is True:
                label.config(text=(players[0]+ " wins"))
                
            elif check_winner() == "Tie":
                label.config(text=("Tie!"))
        else:
            buttons[row][column]['text'] = player

            if check_winner() is False:
                player = players[0]
                label.config(text=(players[0]+" turn"))

            elif check_winner() is True:
                label.config(text=(players[1]+" wins"))

            elif check_winner() == "Tie":
                label.config(text="Tie!")
            
def check_winner():
    #checking all horizontal win conditions
    for row in range(3):
        #we need to check the text of each button in each row here it means somebody won
        if buttons[row][0]['text'] == buttons[row][1]['text'] == buttons[row][2]['text'] != "":
            #optional
            buttons[row][0].config(bg="green")
            buttons[row][1].config(bg="green")
            buttons[row][2].config(bg="green")
            return True
        
    #checking all vertical win conditions
    for column in range(3):
        #we need to check the text of each button in each column here it means somebody won
        if buttons[0][column]['text'] == buttons[1][column]['text'] == buttons[2][column]['text'] != "":
            #optional
            buttons[0][column].config(bg="green")
            buttons[1][column].config(bg="green")
            buttons[2][column].config(bg="green")
            return True
        
    #checking all diagonal win conditions
    #we need to check the text of each button in each diagonal here it means somebody won
    #diagonal 1
    if buttons[0][0]['text'] == buttons[1][1]['text'] == buttons[2][2]['text'] != "":
        #optional
        buttons[0][0].config(bg="green")
        buttons[1][1].config(bg="green")
        buttons[2][2].config(bg="green")
        return True
            
    #diagonal 2
    elif buttons[0][2]['text'] == buttons[1][1]['text'] == buttons[2][0]['text'] != "":
        #optional
        buttons[0][2].config(bg="green")
        buttons[1][1].config(bg="green")
        buttons[2][0].config(bg="green")
        return True
    elif empty_spaces() is False:
        #if there is a tie whole frame will become yellow
        for row in range(3):
            for column in range(3):
                buttons[row][column].config(bg="yellow")
        return "Tie"
    #if there is no winner or no tie
    else:
        return False
        
def empty_spaces():
    spaces = 9
    for row in range(3):
        for column in range(3):
            if buttons[row][column]['text'] != "":
                spaces -= 1
    #no space left here
    if spaces == 0:
        return False
    else:
        return True
def new_game():
    
    global player
    player = random.choice(players)
    label.config(text=player+" turn")

    for row in range(3):
        for column in range(3):
            buttons[row][column].config(text="",bg="#F0F0F0")
#to run window in background
window=Tk()
#setting up the title
window.title("TIC-TAC-TOE")
players = ["X","O"]
#we need select a random player to begin
player = random.choice(players)
#make 3 rows and initialize every row for a time being with zero and put these rows in a 2d list named buttons 
buttons=[[0,0,0], 
         [0,0,0],
         [0,0,0]]
#putting a label to display whose turn it is
label = Label(text=player + " turn" , font=('Monaco',40))
label.pack(side="top")
#creating a reset button to call the new_game function above
reset_button = Button(text="restart", font=('Monaco',20), command =new_game)
reset_button.pack(side="top")
#placing all the buttons inside the frame and adding frame to window then pack the frame
frame= Frame(window)
frame.pack()
#we take 2d lsits of buttons as above and add a button to each spot so use nested for loops here for the above mentioned buttons
for row in range(3):
    for column in range(3):
        buttons[row][column]= Button(frame, text="", font=('Monaco',40), height=2, width=5,
        command=lambda row=row, column=column:  next_turn(row,column))
        buttons[row][column].grid(row=row,column=column)


#to display window
window.mainloop()
```
